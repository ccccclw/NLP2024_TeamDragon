```python
from torchtext.data.metrics import bleu_score

# 1-gram bleu
with open('candidate40.txt', 'r', encoding='utf-8') as f:
    candidate_corpus = [line.strip().split() for line in f]

with open('references correct priming.txt', 'r', encoding='utf-8') as f:
    references_corpus = [line.strip().split() for line in f]

assert len(candidate_corpus) == len(references_corpus), "The number of lines in the two files is different"

bleu_scores = []

for i in range(len(candidate_corpus)):
    candidate = [candidate_corpus[i]]
    reference = [[references_corpus[i]]]
    bleu = bleu_score(candidate, reference)
    bleu_scores.append(bleu)
    print(f'第{i+1}行BLEU分数: {bleu:.4f}')

avg_bleu = sum(bleu_scores) / len(bleu_scores)
print(f'1-gram BLEU Mean: {avg_bleu:.4f}')
```

```python
from torchtext.data.metrics import bleu_score

# 2-gram bleu
with open('candidate.txt', 'r', encoding='utf-8') as f:
    candidate_corpus = [line.strip().split() for line in f]

with open('references correct priming.txt', 'r', encoding='utf-8') as f:
    references_corpus = [line.strip().split() for line in f]

# 确保两个文件行数相同
assert len(candidate_corpus) == len(references_corpus), "The number of lines in the two files is different"

bleu_scores = []

for i in range(len(candidate_corpus)):
    candidate = [candidate_corpus[i]]
    reference = [[references_corpus[i]]]
    bleu = bleu_score(candidate, reference, max_n=2, weights=[0.7, 0.3])
    bleu_scores.append(bleu)
    print(f'第{i+1}行2-gram BLEU分数: {bleu:.4f}')

avg_bleu = sum(bleu_scores) / len(bleu_scores)
print(f'2-gram BLEU Mean: {avg_bleu:.4f}')
```

```python
from torchtext.data.metrics import bleu_score
import csv

with open('candidate.txt', 'r', encoding='utf-8') as f:
    candidate_corpus = [line.strip().split() for line in f]

with open('references.txt', 'r', encoding='utf-8') as f:
    references_corpus = [line.strip().split() for line in f]

assert len(candidate_corpus) == len(references_corpus), "The number of rows in the two files is different"

with open('bleu_scores_correct_translate.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['Row', 'BLEU Score'])

    for i in range(len(candidate_corpus)):
        candidate = [candidate_corpus[i]]
        reference = [[references_corpus[i]]]
        bleu = bleu_score(candidate, reference, max_n=3, weights=[1/3, 1/3, 1/3])
        csvwriter.writerow([i+1, bleu])
```
